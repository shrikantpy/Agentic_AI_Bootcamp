# Sets

## What is a Set in Python?

A set in Python is an unordered collection of unique elements. It is mutable, meaning you can add or remove elements after creation, but the elements themselves must be immutable (e.g., numbers, strings, tuples).

### Key Features:
- **Unordered**: Elements do not have a specific order.
- **Unique**: No duplicate elements are allowed.
- **Mutable**: You can modify the set by adding or removing elements.
- **No Indexing**: You cannot access elements by index like in lists.

### Creating a Set:

````python
my_set = set()
print(f"\ntype of my_set: {type(my_set)}")
print(f"Initial my_set: {my_set}")
````

````python
# Adding a new element to the set
my_set.add("apple")
print(f"my_set after adding 'apple': {my_set}")
````

````python
my_set.add("banana")
my_set.add("orange")
my_set.add("apple")  # Duplicate, won't be added
my_set.add("grape")
my_set.remove("banana")
my_set.discard("kiwi")  # No error if not present
print(f"Final my_set: {my_set}")
# my_set.remove("kiwi")  # This will raise a KeyError if "kiwi" is not present
````

````python
removed_element = my_set.pop()
print(f"my_set after popping an element: {my_set}")
print(f"Popped element: {removed_element}")
````

````python
# Membership test
element_to_check = "orange"
if element_to_check in my_set:
    print(f"{element_to_check} is present in the set.")
else:
    print(f"{element_to_check} is not present in the set.")

element_to_check = "banana"
if element_to_check in my_set:
    print(f"{element_to_check} is present in the set.")
else:
    print(f"{element_to_check} is not present in the set.")
````

````python
# Mathematical set operations
set_a = {"apple", "banana", "cherry"}
set_b = {"banana", "dragonfruit", "elderberry"}
union_set = set_a.union(set_b)
intersection_set = set_a.intersection(set_b)
difference_set = set_a.difference(set_b)
print(f"Union of set_a and set_b: {union_set}")
print(f"Intersection of set_a and set_b: {intersection_set}")
print(f"Difference of set_a and set_b: {difference_set}")  
````

````python
set_a.intersection_update(set_b)
print(f"set_a after intersection_update with set_b: {set_a}")
````

````python
# Symmetric difference
symmetric_difference_set = set_a.symmetric_difference(set_b)
print(f"Symmetric difference of set_a and set_b: {symmetric_difference_set}")
````

````python
# Set methods summary
# add(), remove(), discard(), pop(), clear(), union(), intersection(), difference(), symmetric_difference()

print(set_a.issubset(set_b))
print(set_b.issuperset(set_a))
````

# Dictionary

## What is a Dictionary in Python?

A dictionary in Python is a mutable, unordered collection of key-value pairs. It allows you to store and retrieve data efficiently using unique keys instead of indices like lists.

### Key Features:
- **Keys**: Must be immutable (e.g., strings, numbers, tuples). Keys are unique within a dictionary.
- **Values**: Can be any data type, including lists, other dictionaries, etc.
- **Mutable**: You can add, remove, or modify key-value pairs after creation.
- **Unordered**: Items do not have a specific order (though in Python 3.7+, insertion order is preserved).

### Creating a Dictionary:
```python
# Empty dictionary
my_dict = {}

# Dictionary with initial values
student = {"name": "Alice", "age": 25, "courses": ["Math", "Science"]}
```

### Accessing Values:
Use square brackets with the key:
```python
print(student["name"])  # Output: Alice
```

### Common Operations:
- Add/Update: `student["grade"] = "A"`
- Remove: `del student["age"]`
- Check existence: `"name" in student`
- Get all keys: `student.keys()`
- Get all values: `student.values()`
- Get key-value pairs: `student.items()`

Dictionaries are powerful for mapping relationships and are widely used in Python for data manipulation and configuration.

````python
# Defining an emmpty dictionary
# There are two common ways to define an empty dictionary in Python:

# Method 1: Using curly braces {}
empty_dict = {}
print(f"Empty dictionary using curly braces: {empty_dict}")
print(f"type of empty_dict using curly braces: {type(empty_dict)}")

# Method 2: Using the dict() constructor
empty_dict = dict()
print(f"Empty dictionary using dict(): {empty_dict}")
print(f"type of empty_dict using dict(): {type(empty_dict)}")
````

````python
student = {"Name": "John", "Age": 21, "Courses": ["Math", "CompSci"], "GPA": 4.2}
print(f"Student Dictionary: {student}")
print(f"Name: {student['Name']}")
print(f"Age: {student['Age']}")
print(f"Courses: {student['Courses']}")
print(f"GPA: {student['GPA']}")
````

````python
# Repeated keys in dictionaries silently overwrite values, which can lead to data loss. Always ensure keys are unique.
# Example:
test = {"Name": "Shrikant", "Subject": ["Maths, Science"], "Grade": "A", "Subject": "Computer"}
print(test)
````

````python
# Accessing dictionary elements:
student = {"Name": "John", "Age": 21, "Courses": ["Math", "CompSci"], "GPA": 4.2}
print(f"Student grade is: {student["GPA"]}")
print(f"Student age is : {student['Age']}")
````

````python
# Accessing dictionary values using get() method:
print(student.get("Name"))
print(student.get("Address", "Address not found"))  # Default value if key doesn't exist
print(student.get("Course"))  # None if key doesn't exist
print(student.get("Courses"))
print(student.get("Courses")[0]) # Accessing first course
````

````python
# Modifying dictionary elements
# Modifying existing key name is not possible, but we can achieve it by adding a new key with the desired name and deleting the old key.
student = {"Name": "John", "Age": 21, "Courses": ["Math", "CompSci"], "GPA": 4.2}
student["Age"] = 22  # Modifying existing key value
student["Major"] = "Computer Science"  # Adding a new key-value pair
del student["GPA"]  # Deleting a key-value pair
student["Courses"] = ["Math", "CompSci", "Physics"]  # Modifying the list of courses
student["Courses"].append("Biology")  # Adding a new course to the list
print(f"Modified Student Dictionary: {student}")
````
