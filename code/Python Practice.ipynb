# Sets

```python
my_set = set()
print(f"\ntype of my_set: {type(my_set)}")
print(f"Initial my_set: {my_set}")
```

```python
# Adding a new element to the set
my_set.add("apple")
print(f"my_set after adding 'apple': {my_set}")
```

```python
my_set.add("banana")
my_set.add("orange")
my_set.add("apple")  # Duplicate, won't be added
my_set.add("grape")
my_set.remove("banana")
my_set.discard("kiwi")  # No error if not present
print(f"Final my_set: {my_set}")
# my_set.remove("kiwi")  # This will raise a KeyError if "kiwi" is not present
```

```python
removed_element = my_set.pop()
print(f"my_set after popping an element: {my_set}")
print(f"Popped element: {removed_element}")
```

```python
# Membership test
element_to_check = "orange"
if element_to_check in my_set:
    print(f"{element_to_check} is present in the set.")
else:
    print(f"{element_to_check} is not present in the set.")

element_to_check = "banana"
if element_to_check in my_set:
    print(f"{element_to_check} is present in the set.")
else:
    print(f"{element_to_check} is not present in the set.")
```

```python
# Mathematical set operations
set_a = {"apple", "banana", "cherry"}
set_b = {"banana", "dragonfruit", "elderberry"}
union_set = set_a.union(set_b)
intersection_set = set_a.intersection(set_b)
difference_set = set_a.difference(set_b)
print(f"Union of set_a and set_b: {union_set}")
print(f"Intersection of set_a and set_b: {intersection_set}")
print(f"Difference of set_a and set_b: {difference_set}")  
```

```python
set_a.intersection_update(set_b)
print(f"set_a after intersection_update with set_b: {set_a}")
```

```python
# Symmetric difference
symmetric_difference_set = set_a.symmetric_difference(set_b)
print(f"Symmetric difference of set_a and set_b: {symmetric_difference_set}")
```

```python
# Set methods summary
# add(), remove(), discard(), pop(), clear(), union(), intersection(), difference(), symmetric_difference()

print(set_a.issubset(set_b))
print(set_b.issuperset(set_a))
```

```python
# Counting unique words in a list
test_text = "This is a test text with several words. This text is for testing unique word count."
words = test_text.split()
unique_words = set(words)
unique_word_count = len(unique_words)
print(f"Unique words: {unique_words}")
print(f"Number of unique words: {unique_word_count}")
```

# Dictionary

```python
# Defining an emmpty dictionary
# There are two common ways to define an empty dictionary in Python:

# Method 1: Using curly braces {}
empty_dict = {}
print(f"Empty dictionary using curly braces: {empty_dict}")
print(f"type of empty_dict using curly braces: {type(empty_dict)}")

# Method 2: Using the dict() constructor
empty_dict = dict()
print(f"Empty dictionary using dict(): {empty_dict}")
print(f"type of empty_dict using dict(): {type(empty_dict)}")
```
