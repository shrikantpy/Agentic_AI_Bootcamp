# List practice

### Enter a ***TO-DO*** list

* Enter a TO-DO list to manage tasks

```python
to_do_list = [
    "Learn Pandas",
    "Practice NumPy",
    "Create Visualizations with Matplotlib",
    "Explore Seaborn"
]
```

```python
# Adding a new task to the to-do list
to_do_list.append("Build a Data Science Project")

#Adding another task
to_do_list.append("Study Machine Learning Algorithms")

print("To-Do List:")
for task in to_do_list:
    print(f"- {task}")

# Remove a completed task
to_do_list.remove("Practice NumPy")
print("\nUpdated To-Do List:")
for task in to_do_list:
    print(f"- {task}")

# Check a task in the to-do list
task_to_check = "Learn Pandas"
if task_to_check in to_do_list:
    print(f"\nDon't forget to complete the task: '{task_to_check}'")
else:
    print(f"\nGood job! You have completed the task: '{task_to_check}'")
```

```python
grades = [85, 92, 78, 90, 88]
average_grade = np.mean(grades)
print(f"\nAverage Grade: {average_grade}")

# Adding a new grade
grades.append(95)
new_average_grade = np.mean(grades)
print(f"New Average Grade after adding 95: {new_average_grade}")

#finding highest and lowest grades
highest_grade = np.max(grades)
lowest_grade = np.min(grades)
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```
